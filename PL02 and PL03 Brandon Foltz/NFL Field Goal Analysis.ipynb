```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def sd(x):
    sd = 0 
    mean = x.mean()
    
    for i in x:
        sd = sd + (i-mean)**2
    return np.sqrt(sd/(len(x)-1))
def coeff_of_variation(x):
    s = sd(x)
    m = x.mean()
    return str((s/m)*100) + '%'  
```
