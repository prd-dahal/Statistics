# Plotting Frequency Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
phone_users_name = ['Apple','HTC','LG','Motorola','Samsung']
phone_users_frequency = [48,2,10,4,28]

# df = pd.DataFrame(phone_users)
# df
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(phone_users_name,phone_users_frequency)
ax.set_yticks(np.arange(0,max(phone_users_frequency),5))
ax.set_title('Phone company and users frequency')
```

```python
relative_frequency = []
for i in phone_users_frequency:
    relative_frequency.append(i/np.sum(phone_users_frequency))
print(relative_frequency)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(phone_users_name, relative_frequency)
ax.set_title("Company Name and Relative Frequency")
ax.set_xlabel('Name of the Company')
ax.set_ylabel("Relative Frequency")
```

### Pie Chart

```python
fig  = plt.figure()
ax = fig.add_axes([0,0,1,1])


ax.pie(phone_users_frequency,labels=phone_users_name)
```

## Mean Median Mode (Basic Statistics)

```python
import numpy as np
import scipy
from scipy import stats
data =[1,2,43,23,23,23,23,23,23,23,23,54]
print(np.mean(data))
print(np.median(data))
print(stats.mode(data))

# trimeed mean to  avoid affect of outliers
print(stats.trim_mean(data,0.1))
```
